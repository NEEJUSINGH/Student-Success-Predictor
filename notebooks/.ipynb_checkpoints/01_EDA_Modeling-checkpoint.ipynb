```python
import pandas as pd
df = pd.read_csv("../data/raw/student_performance.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
df["target"] = (df["FinalGrade"] >= 70).astype(int)
df[["FinalGrade", "target"]].head()
```

```python
df["target"].value_counts()
```

```python
features = ['Gender', 'AttendanceRate', 'StudyHoursPerWeek','PreviousGrade', 'ExtracurricularActivities', 'ParentalSupport']
X= df[features]
y= df["target"]
#one hot encode
X_encoded = pd.get_dummies(X, columns = ["Gender", "ParentalSupport"], drop_first = True)
X_encoded.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
# Initialize and train
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# predict on the Test set
y_pred = model.predict(X_test)
```

```python
# Evaluate
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
import joblib
joblib.dump(model, "../models/student_success_mode.pkl")
```
